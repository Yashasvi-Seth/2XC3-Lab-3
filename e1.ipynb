```python
#import statements and setup
import random
from rbt import RBTree
from bst import BST

numOfTrials = 100
length = 10000
```

```python

def run():
    bstTotal = 0
    rbtTotal = 0
    diffTotal = 0

    #calculate total height of the BST and RBT for all trials, and the total difference in height between the two trees
    for i in range(numOfTrials):
        nums = random.sample(range(1, 100001), length)
        bst = BST()
        rbt = RBTree()
        #insert the same random numbers into both trees
        for num in nums:
            bst.insert(num)
            rbt.insert(num)
        bstHeight = bst.get_height()
        rbtHeight = rbt.get_height()

        bstTotal += bstHeight
        rbtTotal += rbtHeight
        diffTotal += (bstHeight - rbtHeight)

    #average each result by dividing the total by the number of trials
    avgHeightBST = bstTotal / numOfTrials
    avgHeightRBT = rbtTotal / numOfTrials
    diffAvg = diffTotal / numOfTrials

    #print the avg results
    print(f"Trials: {numOfTrials}, List Size: {length}")
    print(f"BST avg height: {avgHeightBST}")
    print(f"RBT avg height: {avgHeightRBT}")
    print(f"Avg height difference: {diffAvg}")

#call function
run()


```
