```python
#import statements and setup
import random
from rbt import RBTree
from bst import BST

numOfTrials = 100
length = 10000
```

```python
def run():
    bstTotal = 0
    rbtTotal = 0
    diffTotal = 0

    for i in range(numOfTrials):
        nums = random.sample(range(1, 100001), length)
        bst = BST()
        rbt = RBTree()
        for num in nums:
            bst.insert(num)
            rbt.insert(num)
        bstHeight = bst.get_height()
        rbtHeight = rbt.get_height()

        bstTotal += bstHeight
        rbtTotal += rbtHeight
        diffTotal += (bstHeight - rbtHeight)

    avgHeightBST = bstTotal / numOfTrials
    avgHeightRBT = rbtTotal / numOfTrials
    diffAvg = diffTotal / numOfTrials

    print(f"Trials: {numOfTrials}, List Size: {length}")
    print(f"BST avg height: {avgHeightBST}")
    print(f"RBT avg height: {avgHeightRBT}")
    print(f"Avg height difference: {diffAvg}")

run()


```
