```python
#import statements and setup
import random
import matplotlib.pyplot as plt
from rbt import RBTree
from bst import BST

numOfTrials = 50
length = 5000
swapCounts = [0, 1, 2, 3, 4, 5, 10, 25, 50, 100, 250, 500, 1000, 2500, 5000]
```

```python
#function to apply random swaps to a list
def applySwaps(lst, numSwaps):
    for i in range(numSwaps):
        pos1, pos2 = random.sample(range(len(lst)), 2)
        lst[pos1], lst[pos2] = lst[pos2], lst[pos1]
 

 #function that uses these numbers to build a BST and RBT and returns them
def buildTrees(nums):
    bst = BST()
    rbt = RBTree()
    for num in nums:
        bst.insert(num)
        rbt.insert(num)
    return bst, rbt
 
#main run function
def run():
    avgHeightDiffs = []
    #for every swap, we will run 100 trials and calculate the average height difference between the BST and RBT
    for swaps in swapCounts:
        diffs = []
        #for every trial, create a list of numbers from 1 to 10000, apply swaps on them, and then build a BST and RBT with those numbers and calculate the height difference between them 
        for i in range(numOfTrials):
            nums = list(range(1, length + 1))
            applySwaps(nums, swaps)
 
            bst, rbt = buildTrees(nums)
            diffs.append(bst.get_height() - rbt.get_height())
 
        #calculate the average height difference for this number of swaps and add it to the list of average height differences
        avgHeightDiff = sum(diffs) / numOfTrials
        avgHeightDiffs.append(avgHeightDiff) 
    #plot the number of swaps vs the average difference in height between the BST and RBT
    plt.figure(figsize=(10, 6))
    plt.plot(swapCounts, avgHeightDiffs, marker='o', color='steelblue')
 
    plt.xlabel("Number of Swaps")
    plt.ylabel("Average Height Difference between BST and RBT")
    plt.title("Number of Swaps vs Average Difference in Height for BST and RBT")
    plt.grid(True)
    plt.show()
 
run()
```
