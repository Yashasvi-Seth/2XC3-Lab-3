```python
from xc3t import XCThreeTree
import matplotlib.pyplot as plt
import math 
```

```python
# Build XC3-Trees for degrees 0 to 25 and record height
degrees = [x for x in range(0, 26)]
heights = []
for degree in degrees:
    tree = XCThreeTree(degree=degree)
    heights.append(tree.height())

plt.plot(degrees, heights)
plt.xlabel("Degree i")
plt.ylabel("Height h(i)")
plt.title("Height vs degree of XC3-Tree")
plt.show()
```

```python
# Compare measured heights against formula ceil(i/2)
formula_values = [math.ceil(d / 2) for d in degrees]

plt.plot(degrees, heights, label="h(i) measured")
plt.plot(degrees, formula_values, linestyle="--", label="ceil(i/2)")
plt.xlabel("Degree i")
plt.ylabel("Height h(i)")
plt.title("Height of XC3-Tree vs ceil(i/2)")
plt.legend()
plt.show()

# Verify they match exactly
print(heights == formula_values)
```

## Derived Formula

**h(i) = ⌈i / 2⌉**

The *i*-th child of a degree-*i* root has degree *i* − 2 (for *i* > 2), which is always the tallest child. This gives the recurrence:

h(i) = 1 + h(i − 2), with h(0) = 0 and h(1) = h(2) = 1

Using this formular if we have:

**h(5) = 1 + 1 + 1 = 3**

**h(12) = 1 + 1 + 1 + 1 + 1 + 1 = 6**

**h(9) = 1 + 1 + 1 + 1 + 1 = 5**

Therefore, we can conclude that for every possible i the formula works.

Unrolling by 2 each step yields **h(i) = ⌈i/2⌉**.

Since the number of nodes *n(i)* grows like Fibonacci, and Fibonacci grows exponentially, we have i = O(log n), so **h(i) = O(log n)**.
