```python
from xc3t import XCThreeTree
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Purpose: find pattern in number of nodes vs degree of tree root
degrees = [x for x in range(0, 26)]
num_nodes = []
for degree in degrees:
    tree = XCThreeTree(degree=degree)
    num_nodes.append(tree.nodes())

plt.plot(degrees, num_nodes)
plt.xlabel("Degree of tree root")
plt.ylabel("Number of nodes in tree")
plt.title("Number of nodes vs degree of tree root")
plt.show()
```

```python
# Compare this to fib
def fib(n):
    if n <= 1:
        return n
    a, b = 0, 1
    for _ in range(n):
        a, b = b, a + b
    return a

# plot num nodes vs fibs with different offset
fib_values = [fib(x) for x in degrees]
fib1_values = [fib(x + 1) for x in degrees]
fib2_values = [fib(x + 2) for x in degrees]

plt.plot(degrees, num_nodes, label="Number of nodes in tree")
plt.plot(degrees, fib_values, label="Fib (n)")
plt.plot(degrees, fib1_values, label="Fib (n + 1)")
plt.plot(degrees, fib2_values, label="Fib (n + 2)")
plt.xlabel("i or n")
plt.ylabel("Value")
plt.title("Number of nodes vs Fibonacci numbers")
plt.legend()
plt.show()
```

```python
# Compare each num_nodes with fib (n + 2), return true if all values match
def compare_num_nodes_with_fib(num_nodes, fib_values):
    for num_node, fib_value in zip(num_nodes, fib_values):
        if num_node != fib_value:
            return False
    return True

print(compare_num_nodes_with_fib(num_nodes, fib2_values))
```
